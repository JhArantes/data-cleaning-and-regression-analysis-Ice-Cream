```python
import pandas as pd
```

```python
df = pd.read_excel('exercicio_data_cleaning_RL_Base_Erro.xlsx')
df
```

```python
df['preco'] = (
    df['preco'].astype(str)
    .str.replace('R$', '', regex=False)
    .str.replace(',', '.', regex=False)
    .str.replace(' ', '', regex=False)
    .str.strip())

df['preco'] = pd.to_numeric(df['preco'], errors='coerce')
```

```python
df['preco'].values
```

```python
df['demanda'].values
```

```python
df['demanda'] = df['demanda'].astype(str).str.replace(' un', '', regex=False).str.strip()
```

```python
df['demanda'].values
```

```python
df['demanda'] = pd.to_numeric(df['demanda'], errors='coerce')
```

```python
df['temperatura_c'].values
```

```python
df['temperatura_c'] = df['temperatura_c'].astype(str).str.replace(' C', '', regex=False).str.replace(',', '.', regex=False).str.strip()

df['temperatura_c'].values
```

```python
df['temperatura_c'] = pd.to_numeric(df['temperatura_c'], errors='coerce')
```

```python
df['umidade_pct'].values
```

```python
df['umidade_pct'] = pd.to_numeric(df['umidade_pct'], errors='coerce')

df.loc[(df['umidade_pct'] < 0) | (df['umidade_pct'] > 100), 'umidade_pct'] = None

df['umidade_pct'] = df['umidade_pct'].fillna(df['umidade_pct'].mean())

df['umidade_pct'] = df['umidade_pct'].round(2)
```

```python
df['umidade_pct'].values
```

```python
df['dia_semana'].values
```

```python
mapa_dias = {
    'seg': 1,
    'ter': 2,
    'qua': 3,
    'qui': 4,
    'sex': 5,
    'sab': 6,
    'dom': 7
}

def padronizar_dia(valor):
    if isinstance(valor, str):
        valor = valor.strip().lower()
        return mapa_dias.get(valor, None)
    else:
        return valor

df['dia_semana'] = df['dia_semana'].apply(padronizar_dia)
```
